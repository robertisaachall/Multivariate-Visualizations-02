```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/LSTM-Multivariate_pollution.csv')

air_pollution_important = df[['pollution','dew','temp','press','wnd_dir']]


```

# Python code to visualize 4-variable components via a Pairwise plot. Diagonal Axis is a histogram, Upper Section is KDE, Lower Section is Scatterplot.

```python
plt.figure(figsize=(12,10))
graph = sns.PairGrid(air_pollution_important,hue='wnd_dir')
graph.add_legend()
graph.map_diag(sns.histplot)
graph.map_lower(sns.scatterplot)
graph.map_upper(sns.kdeplot)
plt.show()
```
