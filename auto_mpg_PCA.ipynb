```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA



auto_mpg = pd.read_csv("data/auto-mpg.csv")
features = ['mpg','cylinders','displacement','weight','acceleration'] 

x_features = auto_mpg.loc[:,features].values
y_target = auto_mpg.loc[:,['origin']].values
x_standardized = StandardScaler().fit_transform(x_features)


pca = PCA(n_components = 2)
principal_components = pca.fit_transform(x_standardized)
pca_dataframe = pd.DataFrame(data=principal_components, columns = ['PCA_1','PCA_2'])
pca_dataframe_target = pd.concat([pca_dataframe,auto_mpg[['origin']]],axis = 1)

                                  

```

# Code to visualize PCA components 1 and 2 with seperation based on origin.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
sns.scatterplot(data=pca_dataframe_target,x="PCA_1",y="PCA_2",hue="origin")
```

# Code to visualize other dimensions: MPG and Weight with seperation based on origin

```python
plt.figure(figsize=(12,10))
sns.scatterplot(data=auto_mpg,x="mpg",y="weight",hue="origin")
```
