```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#https://www.kaggle.com/rodolfomendes/abalone-dataset
 
df = pd.read_csv('data/abalone.csv')
```

# Python Code to visualize histogram for the 6-variables. Stacking Used.

```python
abalone_important = df[['Sex','Length','Diameter','Height','Shucked weight','Viscera weight','Shell weight']]

plt.figure(figsize=(12,10))
sns.histplot(data=abalone_important,kde=True,multiple="stack")
```

# Python Code to visualize histogram for the 6-variables. Stacking not used. 

```python
abalone_important = df[['Sex','Length','Diameter','Height','Shucked weight','Viscera weight','Shell weight']]

plt.figure(figsize=(12,10))
sns.histplot(data=abalone_important,kde=True)
```
